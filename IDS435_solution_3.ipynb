### Question 1. Kentwood Electronics

Kentwood electronics is interested in marketing its business in the Chicagoland area.
It is considering three alternatives: TV, radio, and internet.
Kentwood electronics has estimated the number of audience reached per ad, the cost per ad, and the maximum number of ads allowed. This information is reported in the following table:  

|        Component       	    |    Television    	    |    Radio    	    |       Internet    	|
|:--------------------------:	|:--------------------:	|:---------------:	|:------------------:	|
| Audience reached per Ad    	|      100,000         	|    14,000       	|      1        	|
|       Cost per Ad          	|       \$2,000         	|     \$300        	|       \$0.1        	|
|      Maximum usage count     	|         10           	|      20         	|        $\infty$          	|

<br> Your goal is to determine the number of each type of media to use such that the number of audience reached is maximized.
You have to ensure that your solutions adhere to the following constraints:

- The number of each type of media used should not exceed the maximum usage count;
- Total cost of ads should be less than the available budget of $18,000.
- To balance the use of advertising types, television adds may only account for at most 50% of the total budget
- The internet ads should account for at least 10% of the audience.

Please answer the following questions:

### **Question 1-3.**
3. Write Python code to solve the model using Gurobi.



### **Answer.**

```python
from gurobipy import Model, GRB

m = Model("Kentwood_Electronics")

# Define decision variables
x1 = m.addVar(vtype=GRB.INTEGER, name="x1")  # Number of TV ads
x2 = m.addVar(vtype=GRB.INTEGER, name="x2")  # Number of Radio ads
x3 = m.addVar(vtype=GRB.INTEGER, name="x3")  # Number of Internet ads

# Set the objective function
m.setObjective(100000 * x1 + 14000 * x2 + x3, GRB.MAXIMIZE)

# Add constraints
# Media usage constraints
m.addConstr(x1 <= 10, "TV_ads_limit")
m.addConstr(x2 <= 20, "Radio_ads_limit")

# Budget constraint
m.addConstr(2000 * x1 + 300 * x2 + 0.1 * x3 <= 18000, "Budget_limit")

# TV budget constraint (at most 50% of total budget)
m.addConstr(2000 * x1 <= 9000, "TV_budget_limit")

# Internet audience constraint (at least 10% of total audience)
m.addConstr(x3 >= 0.10 * (100000 * x1 + 14000 * x2 + x3), "Internet_audience_requirement")

# Optimize the model
m.optimize()

# Display the optimal solution
if m.status == GRB.OPTIMAL:
    print(f"Optimal number of TV ads: {x1.X}")
    print(f"Optimal number of Radio ads: {x2.X}")
    print(f"Optimal number of Internet ads: {x3.X}")
    print(f"Maximum audience reached: {m.ObjVal}")
```

### **Question 1-6.**

6. **(Graduate Only)** Assume producing TV and Radio ads carries an respective upfront cost of $\$5000$ (TV) and $\$1000$ (Radio). This cost is incurred if at least one ad of the relevant type is aired, but does not increase if more than one ad is run.  

    6.1. Write down a MIP formulation for this modified problem. Be careful not to add any non-linear constraints (besides integrality constraints).

    6.2. Rerun this modified model via Gurobi and report the resulting optimal solution and optimal value.


### **Answer.**
### 6.1 MIP Formulation with Upfront Costs

To incorporate the upfront costs for TV and radio ads into the MIP formulation, we introduce binary variables that indicate whether any ads of a particular type are used.

#### Decision Variables
- \(x1\): The number of television ads.
- \(x2\): The number of radio ads.
- \(x3\): The number of internet ads.
- \(y1\): Binary variable where \(y1 = 1\) if any TV ads are used (\(x1 > 0\)) and \(0\) otherwise.
- \(y2\): Binary variable where \(y2 = 1\) if any Radio ads are used (\(x2 > 0\)) and \(0\) otherwise.

#### Objective Function
Maximize audience reach while considering the upfront costs:
\[ {Maximize} \; 100,000x1 + 14,000x2 + x3 - 5000y1 - 1000y2 \]

#### Constraints
1. Media usage constraints remain the same.
   - \(x1 <= 10\)
   - \(x2 <= 20\)
2. Budget constraint (adjusted for upfront costs):
   - \(2000x1 + 300x2 + 0.1x3 + 5000y1 + 1000y2 <= 18,000\)
3. TV budget constraint (at most 50% of the total budget) remains the same.
   - \(2000x1 + 5000y1 <= 9000\)
4. Internet audience constraint remains the same.
   - \(x3 >= 0.10 X (100,000x1 + 14,000x2 + x3)\)
5. Linking constraints for the binary variables:
   - \(x1 <= 10y1\)
   - \(x2 <= 20y2\)

These linking constraints ensure that y1 and y2 are 1 if there's at least one TV or radio ad, respectively.



```python
from gurobipy import Model, GRB

m = Model("Kentwood_Electronics_Modified")

# Define decision variables
x1 = m.addVar(vtype=GRB.INTEGER, name="x1")  # Number of TV ads
x2 = m.addVar(vtype=GRB.INTEGER, name="x2")  # Number of Radio ads
x3 = m.addVar(vtype=GRB.INTEGER, name="x3")  # Number of Internet ads
y1 = m.addVar(vtype=GRB.BINARY, name="y1")   # Binary variable for TV ads
y2 = m.addVar(vtype=GRB.BINARY, name="y2")   # Binary variable for Radio ads

# Set the objective function
m.setObjective(100000 * x1 + 14000 * x2 + x3 - 5000 * y1 - 1000 * y2, GRB.MAXIMIZE)

# Add constraints
m.addConstr(x1 <= 10, "TV_ads_limit")
m.addConstr(x2 <= 20, "Radio_ads_limit")
m.addConstr(2000 * x1 + 300 * x2 + 0.1 * x3 + 5000 * y1 + 1000 * y2 <= 18000, "Budget_limit")
m.addConstr(2000 * x1 + 5000 * y1 <= 9000, "TV_budget_limit")
m.addConstr(x3 >= 0.10 * (100000 * x1 + 14000 * x2 + x3), "Internet_audience_requirement")
m.addConstr(x1 <= 10 * y1, "Linking_constraint_TV")
m.addConstr(x2 <= 20 * y2, "Linking_constraint_Radio")

m.optimize()

if m.status == GRB.OPTIMAL:
    print(f"Optimal number of TV ads: {x1.X}")
    print(f"Optimal number of Radio ads: {x2.X}")
    print(f"Optimal number of Internet ads: {x3.X}")
    print(f"Maximum audience reached: {m.ObjVal}")
```

This output shows that to maximize the audience reached, the optimal advertising strategy is to run 2 TV ads, 12 radio ads, and 44,000 internet ads, resulting in a maximum audience of 406,000.
